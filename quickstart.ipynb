```python
import requests
from bs4 import BeautifulSoup
web = requests.get("https://vnrvjiet.ac.in/")
```

```python
web.status_code
```

```python
soup = BeautifulSoup(web.content, "html.parser")
print(soup.prettify())
```

```python
soup.title
```

```python
headers = [h.get_text(strip=True) for h in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])]
paragraphs = [p.get_text(strip=True) for p in soup.find_all('p')]
links = [a['href'] for a in soup.find_all('a', href=True)]
```

```python
with open("home.txt", "w", encoding="utf-8") as f:
    f.write("Headers:\n" + "\n".join(headers) + "\n\n")
    f.write("Paragraphs:\n" + "\n".join(paragraphs) + "\n\n")
    f.write("Links:\n" + "\n".join(links))
```
